```python
#Create function to calculate null vars
def print_null_freq(df):
     """
     for a given DataFrame, calculates how many values for 
     each variable is null and prints the resulting table to stdout
     """
     # use melt to convert from wide to long format e.g. table to key, value pair
     df_lng = pd.melt(df)
     null_variables = df_lng.value.isnull()
     return pd.crosstab(df_lng.variable, null_variables)
print_null_freq(train)
```

```python
#Create peak hour feature
train['day'] = train['time'] > 7
train['m_peak'] = (train['time'] > 7) & (train['time'] < 10)
train['n_peak'] = (train['time'] > 15) & (train['time'] < 20)
train
```
