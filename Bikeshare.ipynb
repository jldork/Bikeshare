```python
#Create function to calculate null vars
def print_null_freq(df):
     """
     for a given DataFrame, calculates how many values for 
     each variable is null and prints the resulting table to stdout
     """
     # use melt to convert from wide to long format e.g. table to key, value pair
     df_lng = pd.melt(df)
     null_variables = df_lng.value.isnull()
     return pd.crosstab(df_lng.variable, null_variables)
print_null_freq(train)
```

### Delay impact of weather

```python
# Create feature for days since last rain

# first make feature for change of from weather = 3 to (2, 1)

one_hr = datetime.timedelta(hours = 1)
A = train[['datetime','weather']]
A['datetime'] = A['datetime'].apply(lambda x : datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))

A['datetime'] = A['datetime'] - one_hr
A['weather_prev'] = A['weather']
A.drop('weather', 1, inplace=True)      
train['datetime'] = train['datetime'].apply(lambda x : datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
```

```python
weather = train.merge(A, on='datetime')#[['datetime','weather','weather_prev']]
weather['stopped_rain'] = (weather['weather_prev'] == 3) & ((weather['weather'] == 2) | (weather['weather'] == 2))
weather 
```

```python
rains = weather[weather['stopped_rain'] == True]['datetime']

print list(rains)

zero_hrs = datetime.timedelta(hours=0)

def listalesce(list_obj):
    if not list_obj:
        return [zero_hrs]
    else:
        return list_obj


train['hrs_since_rain'] = train['datetime'].apply(lambda row :
                                                  min(listalesce(filter(lambda x:x > zero_hrs ,row - rains)))
                                                  )/one_hr
rain = train[['hrs_since_rain','time','registered','casual']]
#medians_rain = rain.groupby(['time','hrs_since_rain']).median().reset_index()
#medians_rain
#medians_rain.pivot('time','hrs_since_rain')['registered']
#medians_rain.plot(x='hrs_since_rain', y='registered', kind='scatter')

```
