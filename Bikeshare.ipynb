```python
#Create function to calculate null vars
def print_null_freq(df):
     """
     for a given DataFrame, calculates how many values for 
     each variable is null and prints the resulting table to stdout
     """
     # use melt to convert from wide to long format e.g. table to key, value pair
     df_lng = pd.melt(df)
     null_variables = df_lng.value.isnull()
     return pd.crosstab(df_lng.variable, null_variables)
print_null_freq(train)
```

```python
#Set Up the stuff
import pandas as pd
test = pd.read_csv("Data/test.csv")
train = pd.read_csv("Data/train.csv")
sampleSubmission = pd.read_csv("Data/sampleSubmission.csv")


import matplotlib.pyplot as plt


train['time'] = train['datetime'].str.split(' ').str[1].str[0:2].convert_objects(convert_numeric=True)
```

```python
#Does weather affect ridership of registered/unregistered riders?
plt.figure()
fig,axes = plt.subplots(nrows=4,ncols=2,figsize=(15,16))

train['time'] = train['datetime'].str.split(' ').str[1].str[0:2].convert_objects(convert_numeric=True)

train[(train['workingday']==1) & (train['weather']==1)].plot(x='time',y='registered',kind='scatter', ax=axes[0,0])
train[(train['workingday']==1) & (train['weather']==2)].plot(x='time',y='registered',kind='scatter', ax=axes[0,1])
train[(train['workingday']==1) & (train['weather']==3)].plot(x='time',y='registered',kind='scatter', ax=axes[1,0])
train[(train['workingday']==1) & (train['weather']==4)].plot(x='time',y='registered',kind='scatter', ax=axes[1,1])

#results inconclusive. It does not appear that weather has any effect on registered riders during workdays

train[train['weather']==1].plot(x='time',y='casual',kind='scatter', ax=axes[2,0])
train[train['weather']==2].plot(x='time',y='casual',kind='scatter', ax=axes[2,1])
train[train['weather']==3].plot(x='time',y='casual',kind='scatter', ax=axes[3,0])
train[train['weather']==4].plot(x='time',y='casual',kind='scatter', ax=axes[3,1])
```
